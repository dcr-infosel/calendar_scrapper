```python
from urllib.request import urlopen
import json
import numpy as np
import pandas as pd

# Extraigo data de urls
urls_future = [
"https://www.banxico.org.mx/canales/canal_1_122022_es.json?date=1649782611333",
"https://www.banxico.org.mx/canales/canal_1_112022_es.json?date=1649782611383",
"https://www.banxico.org.mx/canales/canal_1_102022_es.json?date=1649782611426",
"https://www.banxico.org.mx/canales/canal_1_092022_es.json?date=1649782611466",
"https://www.banxico.org.mx/canales/canal_1_082022_es.json?date=1649782611509",
"https://www.banxico.org.mx/canales/canal_1_072022_es.json?date=1649782611557",
"https://www.banxico.org.mx/canales/canal_1_062022_es.json?date=1649782611601",
"https://www.banxico.org.mx/canales/canal_1_052022_es.json?date=1649782611643"]

urls_past=[    
"https://www.banxico.org.mx/canales/canal_1_042022_es.json?date=1649782611688",
"https://www.banxico.org.mx/canales/canal_1_032022_es.json?date=1649782611728",
"https://www.banxico.org.mx/canales/canal_1_022022_es.json?date=1649782611772",
"https://www.banxico.org.mx/canales/canal_1_012022_es.json?date=1649782611821"]


x=np.array([])

for i in urls_future:
    page = urlopen(i)
    html_bytes = page.read()
    html = html_bytes.decode("utf-8")

    # Guardo datos como json
    raw_data=json.loads(html)
    # Acceso a sub campos de interes
    filtered_data=raw_data['webFutureResources']
    # Guardo nombre de campos
    fields=list(filtered_data)


    # Saco de cada lote de fechas, las fechas y. las acomodo en una tupla
    for i in fields:
        x=np.append(x,filtered_data[i][0]['futureResources'])
        #print(x)
    
    x=tuple(x)
    df = pd.DataFrame(x, 
                 columns=['referenceDate', 
                          'title'])
```

```python
df.to_csv("calendario_banxico.csv")
```
